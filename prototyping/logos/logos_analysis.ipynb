#### One logo parties

```python
one_logo_parties = parties_logos[(parties_logos.logo2.isna()) & (parties_logos.logo3.isna())]#.reset_index()
```

```python
one_logo_parties.groupby(by = ['country'], dropna=True).count()
```

```python
one_logo_parties[one_logo_parties.country == 'Sweden']
```

```python
one_logo_parties.drop(one_logo_parties[one_logo_parties.country == 'Sweden'].index, inplace=True)
```

```python
one_logo_parties
```

```python
one_logo_parties.groupby(by = ['country'], dropna=True).count()
```

```python

country_set1 = one_logo_parties.groupby('country').apply(lambda x: x.sample(2)).reset_index(drop=True)
```

```python
country_set1
```

```python
sweden_set = parties_logos[(parties_logos.country == 'Sweden')& parties_logos.logo3.isna()].sample(2).reset_index(drop=True)
```

```python
sweden_set
```

```python
final_logos = pd.concat([country_set1,sweden_set]).reset_index(drop=True)
final_logos
```
